```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
# Creatin a new column of age_group
labels = ['Young Adult', 'Adults', 'Middle-Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age' , 'age_group']].head(10)
```

```python
# Creating a new column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['purchase_frequency_days].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','purchase_frequency_days']].head(10)
```

```python
df.drop(columns=['purchase_frequency_days'],inplace=True)
```
